# **instalation guide**

### **Create Virtual Environment**

```python
#one command installs everything
cd your_project_folder
conda env create -f environment.yml
conda activate chatapp_env 
```

### **Activate the environmenet**

```python
conda activate chatapp_env
# Test everything is installed
python -c "from PyQt6.QtWidgets import QApplication; print('PyQt6 WORKS')"
python -c "import matplotlib; print('Matplotlib WORKS')"


#in case libraries didnt get installed , do it manually
conda activate chatapp_env
conda install pyqt -c conda-forge

##OR THIS this will work definitelyyy
pip install PyQt6
cd server && python main.py
```

### **porject structure**

```python
chatapp/
│
├── environment.yml          # Conda environment (Primary)
├── requirements.txt         # Pip requirements (Fallback)
├── setup_conda.sh           # Conda setup script
├── setup_conda.bat          # Windows Conda setup
├── setup_venv.sh            # Venv fallback script
├── .gitignore
│
├── server/
│   ├── __init__.py          # makes server a package
│   ├── main.py              # Starts GUI
│   ├── config.py            # Host, port, settings
│
│   ├── core/
│   │   ├── __init__.py      # makes core a subpackage
│   │   ├── server.py        # Server class (start/stop/listen)
│   │   ├── client_handler.py# Threaded client handler
│   │   ├── message_protocol.py
│
│   ├── ui/
│   │   ├── __init__.py      # makes ui a subpackage
│   │   ├── main_window.py   # Black monochrome main interface
│   │   ├── sidebar_clients.py
│   │   ├── chat_window.py
│   │   ├── navbar.py
│   │   ├── styles.py
│
│   └── utils/
│       ├── __init__.py      # makes utils a subpackage
│       ├── logger.py
│       ├── helpers.py
│
└── client/
    ├── __init__.py          # makes client a package
    ├── main.py              # Starts GUI
    ├── config.py            # Server IP default, port
    │
    ├── core/
    │   ├── __init__.py      # makes core a subpackage
    │   ├── client_socket.py
    │   ├── receiver_thread.py
    │   ├── message_protocol.py
    │
    ├── ui/
    │   ├── __init__.py      # makes ui a subpackage
    │   ├── connect_window.py
    │   ├── chat_window.py
    │   ├── styles.py
    │
    └── utils/
        ├── __init__.py      # makes utils a subpackage
        ├── logger.py
        ├── helpers.py
```
